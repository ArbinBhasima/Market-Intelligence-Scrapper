```python
from database import engine, text

print("Checking connection to Zion_Database...")

try:
    with engine.connect() as conn:
        # This simple query asks the DB for the number 1
        result = conn.execute(text("SELECT 1")).fetchone()
        
        if result[0] == 1:
            print("✅ CONNECTION SUCCESSFUL!")
            print(f"Connected to: {engine.url.database} on port {engine.url.port}")
            
            # Check if our table exists
            table_check = conn.execute(text(
                "SELECT EXISTS (SELECT FROM information_schema.tables WHERE table_name = 'scraped_reports')"
            )).fetchone()
            
            if table_check[0]:
                print("✅ Table 'scraped_reports' found and ready.")
            else:
                print("⚠️ Connection works, but table 'scraped_reports' is missing!")
                
except Exception as e:
    print("❌ CONNECTION FAILED!")
    print(f"Error Details: {e}")
```
