```python
import helpers
from selenium.webdriver import Remote, ChromeOptions
from bs4 import BeautifulSoup

options = ChromeOptions()
url = "https://www.nrb.org.np/category/annual-reports/?department=red"

sbr_connection = helpers.get_sbr_connection()

with Remote(sbr_connection, options=options) as driver:
    driver.get(url)
    html_source = driver.page_source

    soup = BeautifulSoup(html_source, 'html.parser')
    links = soup.find_all("a")

    post_ids = []

    for link in links:
        href = link.attrs.get("href")

        if href and "risingnepaldaily.com/news/" in href:
            # extract ID from URL
            link_id = href.rstrip("/").split("/")[-1]
            post_ids.append(link_id)

    # remove duplicates
    post_ids = list(set(post_ids))

    print(post_ids)
```

```python
detail_pattern = "https://risingnepaldaily.com/news/{item_id}"
detail_links = [detail_pattern.format(item_id=x) for x in post_ids]
```

```python
for li in soup.find_all('li'):
    a_tag = li.find('a')
    
    if a_tag and a_tag.get("href"):
        print(a_tag.get_text(), "->", a_tag.get("href"))
        
        # get next <li>
        next_li = li.find_next("li")
        
        if next_li:
            next_a = next_li.find("a")
            if next_a:
                print("Next:", next_a.get_text(), "->", next_a.get("href"))
                print("-----")
```

```python
for li in soup.find_all('li'):
    print("----- LI START -----")
    
    # 1. Full text inside <li>
    print("TEXT:", li.get_text(strip=True))
    
    # 2. Attributes of <li>
    print("ATTRS:", li.attrs)
    
    # 3. All <a> tags
    a_tags = li.find_all('a')
    for a in a_tags:
        print("A TAG:", a.get_text(strip=True), "->", a.get("href"))
    
    # 4. All <span> tags
    span_tags = li.find_all('span')
    for span in span_tags:
        print("SPAN:", span.get_text(strip=True))
    
    # 5. Raw HTML (optional, useful for debugging)
    print("HTML:", li)
    
    print("------ LI END ------\n")
```
