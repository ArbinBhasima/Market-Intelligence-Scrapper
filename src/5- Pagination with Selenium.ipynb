```python
import helpers
from selenium.webdriver import Remote, ChromeOptions
from bs4 import BeautifulSoup
import time
from datetime import datetime, timezone

options = ChromeOptions()
max_pages = 10
startday = "2026-03-22"
endday = "2026-03-28"

# start_date = datetime.strptime(startday, "%Y-%m-%d").replace(tzinfo=timezone.utc)
# end_date = datetime.strptime(endday, "%Y-%m-%d").replace(tzinfo=timezone.utc)

# start_str = start_date.strftime("%Y-%m-%d")
# end_str = end_date.strftime("%Y-%m-%d")

now = datetime.now(timezone.utc)
startday = now.strftime("%Y-%m-%d")
endday = now.strftime("%Y-%m-%d")

url_pattern = "https://notifynepal.com/en/commodities/history?start={start}&end={end}&commodity={page}"
# url = "https://notifynepal.com/en/commodities/history"
detail_pattern = "https://notifynepal.com/en/commodities/history?commodity={item_id}"
sbr_connection = helpers.get_sbr_connection()
```

```python
# for page in range(1, 11):
#     print(page)
```

```python
html_datas = []
with Remote(sbr_connection, options=options) as driver:
    for page in range(1, max_pages + 1):
        url = url_pattern.format(start=startday, end=endday, page=page)
        print(page, url)
        driver.get(url)
        time.sleep(2)
        html_source = driver.page_source
        html_datas.append(html_source)
```
