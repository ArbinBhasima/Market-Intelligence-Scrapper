```python
import helpers
from selenium.webdriver import Remote, ChromeOptions

options = ChromeOptions()
url = "https://risingnepaldaily.com/categories/business"
sbr_connection = helpers.get_sbr_connection()
html_source = ""
with Remote(sbr_connection, options=options) as driver:
    driver.get(url)
    html_source = driver.page_source

# html_source to printout
```

```python
detail_patern = "https://risingnepaldaily.com/news/{item_id}"
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html_source, 'html.parser')
# soup.get_text()
```

```python
# html elements find
# soup.find("")
# soup.find_all("a") all anchor tag

# links
links = soup.find_all("a")
for link in links:
#    print(link.get_text())
   # print(link.attrs)
  href = link.attrs.get("href")
  print(href)

```

```python
# post_ids = []
# for link in links:
# # if href has id
#     href = link.attrs.get("href")
#     link_id = None
#     if href.startswith("item?id="):
#         link_id = href.replace("item?id=","")
#     if link_id is not None:
#         print(href, link_id)
#         post_ids.append(link_id)

post_ids = []

# Loop through all <a> tags
for link in links:
    href = link.attrs.get("href")
    if href:  # make sure href exists
        # Check if href matches the news URL pattern
        if "risingnepaldaily.com/news/" in href:
            # Extract the numeric ID at the end
            link_id = href.rstrip("/").split("/")[-1]
            print(href, link_id)
            post_ids.append(link_id)

# Remove duplicates if needed
post_ids = list(set(post_ids))
print("All extracted post IDs:", post_ids)
```

```python
detail_pattern = "https://risingnepaldaily.com/news/{item_id}"
detail_links = [detail_pattern.format(item_id=x) for x in post_ids]
print(detail_links)
```
